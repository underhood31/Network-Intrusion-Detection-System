```
def roc_curve_details(prediction,y_test):
    print("ROC Score : ",roc_auc_score(y_test, prediction))
    fpr, tpr, _ = roc_curve(y_test, prediction)
    plt.clf()
    plt.plot(fpr, tpr)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC curve')
    plt.show()
```
